```python
from pynq import Overlay

ol = Overlay("./conv_test06.bit")
```

```python
dma_0 = ol.axi_dma_0
dma_send_0 = ol.axi_dma_0.sendchannel
dma_recv_0 = ol.axi_dma_0.recvchannel
```

```python
from pynq import allocate
import numpy as np

in_data_size = 28 * 28
out_data_size = 1
in_batch_size = 10000
input_buffer_A = allocate(shape=(in_data_size,), dtype=np.int8)
output_buffer = allocate(shape=(out_data_size,), dtype=np.int8)
```

```python
# imgs = np.load("mnist_test_int8_flat.npy")
# labels = np.load("mnist_test_labels.npy")
imgs = np.load("mnist_test_int8_flat.npy")
labels = np.load("mnist_test_labels.npy")
```

```python
num_correct = 0
for i in range(in_batch_size):
    img = imgs[i]
    label = labels[i]

    input_buffer_A.view(np.int8)[:in_data_size] = img
    
    dma_send_0.transfer(input_buffer_A)
    
    output_buffer = allocate(shape=(out_data_size,), dtype=np.int8)
    dma_recv_0.transfer(output_buffer)
    if(i % 100 == 0):
        print(f'finished {i} inputs')
    
    if (output_buffer[0] == label):
        num_correct = num_correct+1

```

```python
print(f"accuracy: {num_correct/in_batch_size * 100 : .2f}%")
print(f"number of correct output: {num_correct}")
```
